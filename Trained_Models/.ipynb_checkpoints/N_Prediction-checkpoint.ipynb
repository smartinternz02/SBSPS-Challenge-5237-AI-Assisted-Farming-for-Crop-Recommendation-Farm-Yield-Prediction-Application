```python
## Importing the libraries 
import pandas as pd
import numpy as np
import scipy as sp
import xgboost as xgb
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
cropRec = pd.read_csv('../Crop_recommendation.csv')
```

```python
cropRec.head()
```

```python
cropRec.corr()
```

```python
cropRec.isnull().sum()
```

```python
cropRec.drop('P', inplace=True, axis=1)
cropRec.drop('K', inplace=True, axis=1)
cropRec.drop('label', inplace=True, axis=1)
```

```python
cropRec
```

```python
###Choosing the Dependent and Independent variables
# Independent Variables
x=cropRec.iloc[:,[1,2,3,4]]
x
```

```python
# Dependent Variable(target column)
y=cropRec.iloc[:,0]
y
```

```python
## TRAIN TEST SPLIT

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
def hyperParameterTuning(x_train, y_train):
    param_tuning = {
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 5, 7, 10],
        'min_child_weight': [1, 3, 5],
        'subsample': [0.5, 0.7],
        'colsample_bytree': [0.5, 0.8],
        'n_estimators' : [100, 200, 500],
        'objective': ['reg:squarederror']
    }

    xgb_model = xgb.XGBRegressor()

    gsearch = GridSearchCV(estimator = xgb_model,
                           param_grid = param_tuning,                        
                           #scoring = 'neg_mean_absolute_error', #MAE
                           #scoring = 'neg_mean_squared_error',  #MSE
                           cv = 5,
                           n_jobs = -1,
                           verbose = 1)

    gsearch.fit(x_train,y_train)

    return gsearch.best_params_
```

```python
##hyperParameterTuning(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
xgb_model = xgb.XGBRegressor(n_estimators=1000, max_depth=7, subsample=0.7, colsample_bytree=0.8,min_child_weight=1,learning_rate=0.01,objective='reg:squarederror')

xgb_model.fit(x_train, y_train)
print('Accuracy : {}'.format(xgb_model.score(x_test,y_test)))
```

```python
xgb_model.predict(np.array([[20.130175,81.604873,7.628473,262.717340]]))
```
