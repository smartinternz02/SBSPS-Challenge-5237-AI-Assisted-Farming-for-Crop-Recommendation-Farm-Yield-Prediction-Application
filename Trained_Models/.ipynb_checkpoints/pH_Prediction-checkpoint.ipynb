```python
## Importing the libraries 
import pandas as pd
import numpy as np
import scipy as sp
import sklearn as sk
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
import xgboost as xgb
```

```python
cropRec = pd.read_csv('../Crop_recommendation.csv')
```

```python
cropRec.head()
```

```python
cropRec.corr()
```

```python
cropRec.isnull().sum()
```

```python
cropRec.drop('N', inplace=True, axis=1)
cropRec.drop('P', inplace=True, axis=1)
cropRec.drop('K', inplace=True, axis=1)
cropRec.drop('temperature', inplace=True, axis=1)
cropRec.drop('label', inplace=True, axis=1)
```

```python
cropRec
```

```python
###Choosing the Dependent and Independent variables
# Independent Variables
x=cropRec.iloc[:,[0,2]]
x
```

```python
# Dependent Variable(target column)
y=cropRec.iloc[:,1]
y
```

```python
## TRAIN TEST SPLIT

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
## FEATURE SCALING

from sklearn.preprocessing import MinMaxScaler

sc=MinMaxScaler()
```

```python
x_train=sc.fit_transform(x_train)
```

```python
x_test=sc.transform(x_test)
```

```python
from sklearn.metrics import mean_squared_error
xgb_model = xgb.XGBRegressor(objective="reg:squarederror", random_state=1)

xgb_model.fit(x_train, y_train)

y_pred = xgb_model.predict(x_test)

mse=mean_squared_error(y_test, y_pred)

print(np.sqrt(mse))

xgb_model.predict(np.array([[70.11713476,231.3843163]]))
```

```python
#Saving our model into a file
import pickle
pickle.dump(xgb_model,open('pH.pkl','wb'))
```
